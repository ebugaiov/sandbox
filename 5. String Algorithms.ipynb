```python
def is_anagram(str1: str, str2: str) -> bool:
    s1 = str1.replace(' ', '').lower()
    s2 = str2.replace(' ', '').lower()
    if sorted(s1) == sorted(s2):
        return True
    return False

assert is_anagram('Hello', 'Olleh') == True
```

```python
def is_palindrome(string: str) -> bool:
    if string.lower() == string[::-1].lower():
        return True
    return False

assert is_palindrome('Helleh') == True
```

```python
from typing import Optional

def get_last_digit(string: str) -> Optional[int]:
    temp = [int(item) for item in string if item.isdigit()]
    if len(temp) > 0:
        return temp[-1]
    return

assert get_last_digit('hello 1 or 2 and...') == 2
assert get_last_digit('hello') == None
```

```python
import string

def shifr(msg: str, shift: int) -> str:
    lower_case = string.ascii_lowercase
    upper_case = string.ascii_uppercase
    encrypt = ''
    for item in msg:
        if item in lower_case:
            new_index = (lower_case.index(item) + shift) % 26
            encrypt += lower_case[new_index]
        elif item in upper_case:
            new_index = (upper_case.index(item) + shift) % 26
            encrypt += upper_case[new_index]
        else:
            encrypt += item
    return encrypt

assert shifr('abc', 1) == 'bcd'
```
